```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def drawBlackSquare(image, location, size):

    x,y = location
    w,h = size

    for row in range(h):
        for col in range(w):
            image[row+y][col+x] = [0,0,0]

    return image

imageArray = np.array(plt.imread('filename.jpg'), dtype=np.uint8)

imArray = drawBlackSquare(imageArray, (0,0), (30,30))

fig = plt.figure()
plt.imshow(imageArray)
```
